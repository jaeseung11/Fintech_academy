### 3) 사이킷런 변환기

특성공학에서는 데이터의 특성들을 이용해서 새로운 특성을 만든다고 했습니다. 그렇다면 일일히 모두 손으로 계산해가면서 특성을 만들어야 할까요?

242F904A549F807530.jfif

아닙니다. 사이킷런을 이용해서 자동으로 특성을 만들수 있습니다.

```python
#sklearn.preprocessing에서  PolynomialFeatures를 사용합니다.
from sklearn.preprocessing import PolynomialFeatures
```

```python
#poly라고 변수의 특성을 자동으로 만들어 내는 함수를 만듭니다.
poly = PolynomialFeatures()

#여기에 2,3을 입력했습니다.
poly.fit([[2, 3]])

#2,3을 입력해서 나온 특성 값들을 보여주게 합니다. 
print(poly.transform([[2, 3]]))
```

```python
#실제 이 값은 무게 = a x 길이 + b x 높이 + c x 두께 + d x 1 꼴입니다. 그런데 여기 d값(절편)은 큰 의미가 없기 때문에 off합니다. 
#bias 옵션을 off합니다.

poly = PolynomialFeatures(include_bias=False)
poly.fit([[2, 3]])
print(poly.transform([[2, 3]]))
```

1이 빠진 것을 알 수 있습니다.

```python
#이제 이렇게 특성변수를 만드는 함수에 실제 생선데이터를 집어넣겠습니다.

poly = PolynomialFeatures(include_bias=False)

poly.fit(train_input)
train_poly = poly.transform(train_input)
```

```python
print(train_input.shape)
```

```python
print(train_poly.shape)
```

```python
#값을 보겠습니다. 3개변수(길이, 너미, 두께)->9개변수로 확장됐군요 
print(train_poly)
```

```python
#실제로 어떻게 특성함수를 만들었는지 보도록 하겠습니다.
poly.get_feature_names()
```

```python
#중요 : 테스트 세트(실제 정확도를 테스트할 데이터도 훈련 세트처럼 변환해야 합니다.
test_poly = poly.transform(test_input)
```

## 릿지 모형



*   규제가 있는 선형 회귀모형으로 선형 모델의 계수를 작게 만들어 과대 적합을 완화시킴 
*   항목 추가



적합도1.jpg

```python
from sklearn.linear_model import Ridge

ridge = Ridge()
ridge.fit(train_scaled, train_target)
print(ridge.score(train_scaled, train_target))
```

```python
print(ridge.score(test_scaled, test_target))
```

```python
import matplotlib.pyplot as plt

train_score = []
test_score = []
```

```python
alpha_list = [0.001, 0.01, 0.1, 1, 10, 100]
for alpha in alpha_list:
    # 릿지 모델을 만듭니다
    ridge = Ridge(alpha=alpha)
    # 릿지 모델을 훈련합니다
    ridge.fit(train_scaled, train_target)
    # 훈련 점수와 테스트 점수를 저장합니다
    train_score.append(ridge.score(train_scaled, train_target))
    test_score.append(ridge.score(test_scaled, test_target))
```

```python
#특성공학 정도가 높아 질수록(alpha) 항상 적합도가 올라가는 것이 아니다. 테스트(검증) 세트에서는 최적의 적합도 값이 존재한다.
plt.plot(np.log10(alpha_list), train_score)
plt.plot(np.log10(alpha_list), test_score)
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.show()
```

(참고)

* 하이퍼 파라미터는 머신러닝 알고리즘이 학습하지 않는 파라미터로, 사전에 별도로 지정해야힘

* 여기서 알파는 대표적인 규제 강도 파라미터임

```python
ridge = Ridge(alpha=0.1)
ridge.fit(train_scaled, train_target)

print(ridge.score(train_scaled, train_target))
print(ridge.score(test_scaled, test_target))
```

```python
train_score = []
test_score = []

alpha_list = [0.001, 0.01, 0.1, 1, 10, 100]
for alpha in alpha_list:
    # 라쏘 모델을 만듭니다
    lasso = Lasso(alpha=alpha, max_iter=10000)
    # 라쏘 모델을 훈련합니다
    lasso.fit(train_scaled, train_target)
    # 훈련 점수와 테스트 점수를 저장합니다
    train_score.append(lasso.score(train_scaled, train_target))
    test_score.append(lasso.score(test_scaled, test_target))
```

```python
#특성공학 정도가 높아 질수록(alpha) 항상 적합도가 올라가는 것이 아니다. 어느 지점에서는 급락한다.

plt.plot(np.log10(alpha_list), train_score)
plt.plot(np.log10(alpha_list), test_score)
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.show()
```
